```python
def callGrasshopper(path,parameters):
    decodedScript = loadGrasshopperScript(path)
    hopsValues = convertToValues(parameters)
    response = requests.post(compute_rhino3d.Util.url + "grasshopper", json={
        "algo": decodedScript,
        "pointer": None,
        "values": hopsValues
    })
    if response.ok:
        contentJson = response.content.decode("utf-8")
        content= json.loads(contentJson)
        values = content["values"]
        return values
    else:
        return response.reason

def loadGrasshopperScript(path):
    gh_data = open(path, mode="r", encoding="utf-8-sig").read()
    data_bytes = gh_data.encode("utf-8")
    encoded = b64encode(data_bytes)
    decoded = encoded.decode("utf-8")
    return decoded

typeMapping = {
    str: "System.String",
    float: "System.Double",
    int: "System.Integer",
    bool: "System.Boolean",
    float64: "System.Double",
    int32: "System.Integer",
    bool_: "System.Boolean",
}
dataConversion = {
    int32 : lambda x: int(x),
    bool_ : lambda x: bool(x)
}

def convertToValues(parameters):
    hopsValues = []
    if type(parameters) is not dict:
        return "Parameters have to be a dictionary"
    for kw in parameters:
        values = parameters[kw]
        if type(values) != list:
            values=[values]
        hopsValues.append(listToHopsValue(kw,values))
    return hopsValues


def listToHopsValue(parameterName, valueList,grasshopperPath="0"):
    return {
        "ParamName": parameterName,
        "InnerTree": {
            "{ "+ grasshopperPath +"; }": [ {
                    "type":  typeMapping[type(v)],
                    "data": dataConversion.get(type(v),lambda x:x)(v)
                } for v in valueList]
        }
    }

def saveResults(values):
    #Save results
    model = File3dm()
    for val in values:
        paramName = val['ParamName']
        innerTree = val['InnerTree']
        isLayerInitialized=False
        for key, innerVals in innerTree.items():
            for innerVal in innerVals:
                #if innerVal['type'] != "System.String":
                try:
                    if not isLayerInitialized:
                        layerIndex = model.Layers.AddLayer(paramName,(0,0,0,255))
                        isLayerInitialized=True
                    data = json.loads(innerVal['data'])
                    decodedData = CommonObject.Decode(data)
                    objectAttributes = ObjectAttributes()
                    objectAttributes.LayerIndex = layerIndex
                    model.Objects.Add(decodedData,objectAttributes)
                except:
                    pass
    #model.Write(simulationResultsDirectory + '\\model.3dm')

parameters = designSpaceMatrix.iloc[0].to_dict()

#Breaks with lists
#parameters['Output-Names']=["Heating Coil Heating Rate","Heating Coil Electricity Rate"]
# parameters['Report-Names']=reportNames
# result = callGrasshopper(r"C:\Git\Work\FOR2363\Parametric-H-Shape-Office-Building-Model\src\Director.ghx",parameters)
# path = ""
# for output in result:
#     if output['ParamName']=="SQL-FilePath":
#         path=output['Inner-Tree']
# result
# path
```
